```python
def sample_method(in0, in1):
    return (in0+in1)

print(sample_method(1, 2))
print(sample_method('a', 'b'))
```

```python
def sample_method(in0, in1, *args):
    print(in0, in1, args[2])

a = [1, 2, 3,4]
sample_method('a', 'b', *a)
```

```python
def sample_method(in0, in1, **kwargs):
    if 'has_kwargs' in kwargs:
        print(in0, in1, kwargs['has_kwargs'])

b = dict(s0=0, s1=1, has_kwargs=True)
sample_method('a', 'b', **b)
```

```python
class Sample():
    def __init__(self, arg0, arg1):
        self.arg1 = arg1
        self.arg0 = arg0
    
    def set_attr(self, arg0, arg1):
        self.arg1 = arg1
        self.arg0 = arg0

sample = Sample('a', 'b')
print(sample.arg0, sample.arg1)
```

```python
sample.set_attr('c', 'd')
print(sample.arg0, sample.arg1)
```

```python
from PIL import Image
```

```python
image = Image.open('red-fox.jpg')
```

```python
type(image)
```

```python
blured = cv2.GaussianBlur(image, (15, 15), 7)
cv2.imwrite('blured-fox.jpg', blured)
plt.imshow(cv2.cvtColor(blured, cv2.COLOR_BGR2RGB))
```

```python
resized = cv2.resize(image, (750, 500), interpolation=cv2.INTER_AREA)
cv2.imwrite('resized-fox.jpg', resized)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.scatter([200], [100], c='red')
```

```python
center_coordinates = (200, 100)
redius = 10
color = (0, 0, 255)
thickness = 5
resized = cv2.circle(resized, center_coordinates, redius, color, thickness)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```
